```python
import pandas as pd
import seaborn as sns
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
iris=sns.load_dataset("iris")
#df=tips.copy()

iris.head()
```

```python
tips['smoker'].value_counts().plot(kind='pie')
```

univriate analysis on numerical data

```python
import matplotlib.pyplot as plt
plt.hist(tips['tip'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.boxplot(tips['total_bill'])
```

**bivariate work on two coulumn of table and multivariate works on multiple column of table**

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

```python
sns.barplot(x='smoker', y='total_bill', data=tips)
plt.title('Average Total Bill by Smoker Status')
plt.xlabel('Smoker')
plt.ylabel('Average Total Bill')
plt.show()
```

```python
titanic=pd.read_csv("train.csv")
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic,color='orange',hue=titanic['Sex'])
#sns.barplot(titanic['Pclass'],titanic['Fare'],hue=titanic['Sex'])
plt.title('Average Fare by Passenger Class   ')
```

**boxplot(numerical-categorical)**

```python
sns.boxplot(x='Sex', y='Age', data=titanic,hue=titanic['Survived'])
```

```python
sns.distplot(titanic['Age'],color='orange')
#sns.distplot(titanic['Pclass'],color='orange')
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False,color='blue')
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False,color='orange')
```
